```python
def check_df(dataframe, head=5):
    print('########### head #########')
    print(dataframe.head(head))
    print('########### shape #########')
    print(dataframe.shape)
    print('######### dtypes #########')
    print(dataframe.dtypes)
    print('########## describe ##########')
    print(dataframe.describe().T)
    print('########### isnull ###########')
    print(dataframe.isnull().values.any())
    print('########### columns ##########')
    print(dataframe.columns)
    print('############ tail ##########')
    print(dataframe.tail(head))
    
    
check_df(df)
```

```python
# Adım 2: Numerik ve kategorik değişkenleri yakalayınız.
def grab_col_name(dataframe, cat_th=5, car_th=10):
    # Category
    cat_cols=[col for col in dataframe.columns if dataframe[col].dtypes=='O']
    num_but_cat=[col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes !='O']
    cat_but_car=[col for col in dataframe.columns if dataframe[col].nunique() > car_th and dataframe[col].dtypes=='O']
    cat_cols+=num_but_cat
    cat_cols=[col for col in cat_cols if col not in cat_but_car]
    
    # Number
    num_cols=[col for col in dataframe.columns if dataframe[col].dtypes !='O']
    num_cols = [col for col in num_cols if col not in num_but_cat]
    
    
    print(f'Gozlem sayisi {dataframe.shape[0]}')
    print(f'Deyisken sayisi {dataframe.shape[1]}')
    print(f'categorik deyiskenler {len(cat_cols)}')
    print(f'number but categorik {len(num_but_cat)}')
    print(f'cardinal {len(cat_but_car)}')
    print(f'number {len(num_cols)}')
    
    return cat_cols, num_but_cat, cat_but_car ,num_cols   
```

```python
cat_cols, num_but_cat, cat_but_car ,num_cols=grab_col_name(df)
```

```python
def category_sum(dataframe, col_name):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                       'Ratio': 100*dataframe[col_name].value_counts()/len(dataframe)}))
    print('###########################################')
```

```python
for col in cat_cols:
    category_sum(df,col)
```

```python
def numerical_sum(dataframe,num_col):
    quantiles=[0.05,0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90,0.95,0.99]
    print(dataframe[num_col].describe(quantiles))
    print('###########################################')
```

```python
for col in num_cols:
    numerical_sum(df,col)
```

```python
for col in cat_cols:
    numerical_sum(df,col)
```

```python
def target_sumarry_with_mean(dataframe, target, numerical_col):
    print(dataframe.groupby(target).agg({numerical_col: 'mean'}))
```

```python
for col in num_cols:
    target_sumarry_with_mean(df,'sales', col)
```
